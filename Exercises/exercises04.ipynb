## 5. Dice rolls convergence (**)

Simulate: 

&nbsp; a) &nbsp; 100 dice rolls and count the number of outcome six. (*)

&nbsp; b) &nbsp; 10, 100, 1000, 10000, 100000, 1000000 dice rolls. Count the number of outcome six in each simulation and store it in a list. Compute the probability of outcome six in each simulation. (**)

&nbsp; c) &nbsp; Use matplotlib to plot this list. (*)


b)

<img align="left" src="../assets/probability_six_graph.png" width="300"/>

</details>

````python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt

def countSixes(rollsNumber):
    diceRolls = [rnd.randint(1, 6) for x in range(rollsNumber)]
    six = [x for x in diceRolls if x == 6]
    return len(six)

def probability(a, b):
    prob = [np.round((x/y), 4) for x, y in zip(a, b)]
    return prob

listOfSix = []
listOfRolls = []

for i in range(1, 7):
    rolls = int(np.pow(10, i))
    numSixes = countSixes(rolls)
    listOfSix.append(numSixes)
    listOfRolls.append(rolls)
prob = probability(listOfSix, listOfRolls)

plt.plot(listOfRolls, prob) 
plt.xscale("log")
plt.title("Probability of getting six")
plt.xlabel("Number of rolls")
plt.ylabel("Probability of getting six")

````

For $d_i < 1$ plot the point with one color, and for $d_i > 1$ plot the point in another color, $i = \{1,\ldots, 5000\}$.  (**)

&nbsp; b) Calculate the fraction between number of inner points and number of outer points. What do you this will converge to when number of points reaches infinity? (**)

````python
import numpy as np
import random as rnd
import matplotlib.pyplot as plt

x = np.array([rnd.uniform(-1, 1) for x in range(5000)])
y = np.array([rnd.uniform(-1, 1) for x in range(5000)])

mask1 = np.sqrt(np.pow(x, 2) + np.pow(y, 2)) < 1
mask2 = np.sqrt(np.pow(x, 2) + np.pow(y, 2)) > 1

innerPoints = len([i for i in mask1 if i == True])
outerPoints = len([i for i in mask2 if i == True])

fraction = innerPoints/ (outerPoints + innerPoints)
theorethicalFraction = np.pi/4

print(f"In theory {theorethicalFraction:.2%} of points will end up in the cirkle")
print(f"In this case {fraction:.2%} of points ended up in the cirkle")

plt.figure(figsize= (6, 6))
plt.title("Monte Carlo simulation")
plt.scatter(x[mask1], y[mask1], color = "#FF1100", s=10, alpha= 0.5)
plt.scatter(x[mask2], y[mask2], color = "#0000FF", s=10, alpha= 0.5)
plt.xlabel("X")
plt.ylabel("Y")
````

## 7. A cute rabbit among two ferocious snakes (**)

There are three identical doors, behind two of them, there are ferocious snakes and behind one of them is a cute little rabbit. Your aim is to pick a door and get the cute rabbit. You start picking one door randomly and suddenly another door opens and snake slithers away *pssss*. Now you have a choice to either stay with your door or change to the remaining door. 

<img src="../assets/snakes_rabbits.png" width = 200>

&nbsp; a) What is your choice and why? (**)

&nbsp; b) Randomly assign which door has the rabbit, and randomly pick an initial door. Now simulate 10, 100, 1000, 10000, 100000, 1000000 times for staying and for switching doors. Calculate the proportion of finding the rabbits for each set of simulations. Plot the results (**)


<details>

<summary>Hint</summary>

a) 

Try to reason using probability.


</details>

<details>

<summary>Answer</summary>


a) 

You will simulate the answer in b).

b) 

<img src="../assets/mh_simulations.png" width = 300>



</details>

````python
#b)
import random as rnd
import matplotlib.pyplot as plt

def shuffleDoors():
    animals =["snake", "snake", "rabbit"]
    rnd.shuffle(animals)
    chosenDoor, rDoor1, rDoor2 = animals
    remainingDoors = [rDoor1, rDoor2]
    rnd.shuffle(remainingDoors)
    return chosenDoor, remainingDoors


numberOfSimulations = [10**i for i in range(1, 7)]
ProbabilityStaying = []
ProbabilityChanging = []

for i in numberOfSimulations:
    resultStaying = []
    resultChanging = []

    for j in range(i):
    
        chosenDoor, remainingDoors = shuffleDoors()
        rDoor1, rDoor2 = remainingDoors
        

        for door in remainingDoors:
            if door == "snake":
                if chosenDoor == "rabbit":
                    answer = True
                    resultStaying.append(answer)
                else:
                    answer = False
                    resultStaying.append(answer)
                break
    
        for door in remainingDoors:
            if door == "snake":
                if door == rDoor1:
                    chosenDoor = rDoor2
                elif door == rDoor2:
                    chosenDoor = rDoor1

                if chosenDoor == "rabbit":
                    answer = True
                    resultChanging.append(answer)
                else:
                    answer = False
                    resultChanging.append(answer)
                break        

    ProbabilityStaying.append(sum(resultStaying)/i)
    ProbabilityChanging.append(sum(resultChanging)/i)

plt.figure(figsize= (6, 6))
plt.plot(numberOfSimulations, ProbabilityStaying, label= "Staying", marker="o", color= "#ff0000")
plt.plot(numberOfSimulations, ProbabilityChanging, label= "Changing", marker="o", color="#0000ff")
plt.xlabel("Number of simulations")
plt.ylabel("Probability of finding the rabbit")
plt.legend()
plt.grid(True)
plt.xscale("log")

````
